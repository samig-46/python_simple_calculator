Calculator

```python
# libraries
import ipywidgets as widgets
from IPython.display import display

# functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Error: Division by zero"

def on_calculate(change):
    a = num1.value
    b = num2.value
    op = operation.value
    if op == '+':
        res = add(a, b)
    elif op == '-':
        res = subtract(a, b)
    elif op == '*':
        res = multiply(a, b)
    elif op == '/':
        res = divide(a, b)

    result.value = f"Result: {res}"
num1 = widgets.FloatText(description='Number 1:', value=0)
num2 = widgets.FloatText(description='Number 2:', value=0)

operation = widgets.Dropdown(
    options=[
        ('Addition', '+'),
        ('Subtraction', '-'),
        ('Multiplication', '*'),
        ('Division', '/')
    ],
    value='+',
    description='Operation:'
)

result = widgets.Label(value='Result: ')

num1.observe(on_calculate, names='value')
num2.observe(on_calculate, names='value')
operation.observe(on_calculate, names='value')
display(num1, num2, operation, result)
```
